```python
import pandas as pd
import numpy as np

df = pd.read_csv('simple_kaggle_dataset.csv')

df.head()
```

```python
df.info()
df.describe()
```

```python
#Faturamento

df['Revenue'] = df['Price'] * df['Quantity_Sold']
df.head
```

```python
#Faturamento Total
df['Revenue'].sum()
```

```python
#Faturamento Total
df.groupby('Category') ['Revenue'].sum().sort_values(ascending=False)
```

```python
#Quantidade de Vendas por Categoria
df.groupby('Category') ['Quantity_Sold'].sum().sort_values(ascending=False)
```
